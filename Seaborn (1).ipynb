- pandas.pivot(index, columns, values) function produces pivot table based on 3 columns of the DataFrame. Uses unique values from index / columns and fills with values.

    * Parameters:
        * index[ndarray] : Labels to use to make new frame’s index
        * columns[ndarray] : Labels to use to make new frame’s columns
        * values[ndarray] : Values to use for populating new frame’s values
 
    * Returns: Reshaped DataFrame
    * Exception: ValueError raised if there are any duplicates.

```python
df.pivot(index = 'Region', columns = 'Type')  # not mentionaing value means--> printing whole dataset
```

```python
df.pivot(index = 'Region', columns= 'Type', values= ['Small Bags', 'Large Bags'])  # pass columns name to be taken 
```

```python
df.pivot(index = 'Region', columns= 'Type')[['Small Bags', 'Large Bags']]   # we can pass this way also
```

- A rug plot is a plot of data for a single quantitative variable, displayed as marks along an axis. It is used to visualise the distribution of the data. 
- As such it is analogous to a histogram with zero-width bins, or a one-dimensional scatter plot

```python
sns.rugplot(pokemon['Attack Point'], 
           height = 0.5, 
           axis = 'x')    # height is given big so that line should be big.....ans seen clearlly
```

```python
# Categorical Plot - swarmplot, barplot, boxplot, violin plot, countplot, catplot

tips = sns.load_dataset('tips')
tips.head()


```

```python
plt.subplots(figsize = (10, 5))
sns.swarmplot(x = 'day',
             y = 'tip',
             data = tips)
```

```python
plt.subplots(figsize = (10, 5))
sns.swarmplot(x = 'day',
             y = 'tip',
             data = tips,
             hue = 'sex')
```

```python
plt.subplots(figsize = (15, 5))
sns.boxplot(x = 'day',
             y = 'tip',
             data = tips, hue = 'size')
```

```python
plt.subplots(figsize = (15, 5))
sns.violinplot(x = 'day',
             y = 'tip',

               data = tips)
```

```python
plt.subplots(figsize = (15, 5))
sns.barplot(x='day',
            y='tip',
            data=tips, hue = 'size')
```
