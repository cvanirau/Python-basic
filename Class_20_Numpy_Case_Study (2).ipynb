```python
# Looping over the image 

# img.shape ---> (533,800,3)  
# [
#     [
#         [],[],[],[]
#     ],
#     [
#         [],[],[],[]
#     ],
#     [
#         [],[],[],[]
#     ],
#     [
#         [],[],[],[]
#     ]
# ]

index = []
img_temp = img.copy()

for i in range(img.shape[0]):
    if(i % 2 == 0):
        index.append(True)
        for j in range(img.shape[1]):
            img_temp[i,j,(1,2)] = 0  # (1,2) points to G B
    else:
        index.append(False)

plt.figure(figsize=(8,8))  # configuring the plot to visualize the image
plt.imshow(img_temp[index])  # ploting the image to the console
```
