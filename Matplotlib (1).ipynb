- The subplot() function takes three arguments that describes the layout of the figure.
- The layout is organized in rows and columns, which are represented by the first and second argument.
- The third argument represents the index of the current plot.
    * plt.subplot(1, 2, 1)
        * the figure has 1 row, 2 columns, and this plot is the first plot.
    * plt.subplot(1, 2, 2)
        * the figure has 1 row, 2 columns, and this plot is the second plot.
- You can add a title to each plot with the title() function:
- You can add a title to the entire figure with the suptitle() function:

```python
plt.subplot(2, 1, 1)
plt.plot(arr1, arr2)
plt.xlabel('Arr1')
plt.ylabel('Arr2')
plt.title('Arr1 vs Arr2')
plt.subplot(2, 1, 2)
plt.plot(arr2, arr1)
plt.xlabel('Arr2')
plt.ylabel('Arr1')
plt.title('Arr2 vs Arr1')
plt.tight_layout()
plt.show()
```

```python
ar1= np.arange(-6,6, 1)
ar2= ar1**3

plt.subplot(2,2,1)
plt.plot(ar1, ar2, 'g')
plt.xlabel('ar1')
plt.ylabel('ar2')
plt.title('ar1 vs ar2')

plt.subplot(2,2,2)
plt.plot(ar2, ar1, 'r')
plt.xlabel('ar2')
plt.ylabel('ar1')
plt.title('ar2 vs ar1')

plt.subplot(2,2,3)
plt.plot(arr1, arr2, 'y')
plt.xlabel('arr1')
plt.ylabel('arr2')
plt.title('arr1 vs arr2')

plt.subplot(2,2,4)
plt.plot(arr2, arr1, 'b')
plt.xlabel('arr2')
plt.ylabel('arr1')
plt.title('arr2 vs arr1')

plt.suptitle('*** Illustration of array in different arrays ***')
plt.tight_layout()
plt.show()
```

```python
# subplots() manages the axes automatically

fig, axes = plt.subplots()
axes.plot(arr1, arr2, 'g')
axes.plot(ar1, ar2*2, 'r')
axes.plot(arr1, arr2*4, 'y')
plt.show()

```

```python
# subplots() manages the axes automatically

fig, axes = plt.subplots(nrows = 2, ncols = 2)
axes[0][0].plot(arr1, arr2, 'r')
axes[0][1].plot(arr2, arr1, 'g')
axes[1][0].plot(ar2, ar1+10, 'b')
axes[1][1].plot(ar1, ar2*2, '')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (10, 5), dpi = 100)
axes[0][0].plot(arr1, arr2, 'r')
axes[0][1].plot(arr2, arr1, 'g')
axes[1][0].plot(ar2, ar1-10, 'b')
axes[1][1].plot(ar1, ar2+10, '')
plt.tight_layout()
plt.show()
```

```python
# Write a function that uses a default value.     
            # In - introduce('Lim', 20)
            # Out - 'My name is Lim. I am 20 years old.'
            # In - introduce('Ahmad')
            # Out - 'My name is Ahmad. My age is secret.'
```

```python
def introduce(name, age = 'secret.'):
    print(f'My name is {name}. My age is {age}')
    
introduce('Lim', 20)
```

```python
# Write a function, is_vowel that returns the value true if a given character is a vowel, 
# and otherwise returns false. 
# Write another function main, in main() function accept a string from user and count number 
# of vowels in that string.
```

```python
vowel = 'aeiouAEIOU'
def is_vowel(char):
    return True if char in vowel else False

def main(my_str):
    count_vowels = 0
    for i in my_str:
        if is_vowel(i):
            count_vowels+=1
    return count_vowels

main('Manoeuver')
```
