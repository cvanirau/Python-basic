```python
ny_weather_df.fillna(method = 'ffill') # forward fill --> filling from previous data 
```

```python
ny_weather_df.fillna(method = 'bfill') # backward fill- filling from next data
```

```python
# A binary file "Book.dat" has structure [BookNo, Book_Name, Author, Price].
# i. Write a user defined function createFile() to input data for a record and add to Book.dat.
# ii. Write a function countRec(Author) in Python which accepts the Author name as parameter 
# and count and return number of books by the given Author are stored in the binary file "Book.dat"
```

```python
import pickle
def createFile():
    f = open('Book.dat', 'wb')
    no_entries = int(input('Enter the no of entries you want to make: '))
    for i in range(no_entries):
        book_no = eval(input('Enter the book No: '))
        book_name = eval(input('Enter the book Name: '))
        author_name = eval(input('Enter the Author Name: '))
        price = eval(input('Enter the book Price: '))
        pickle.dump([book_no, book_name, author_name, price], f)
        
createFile()
```

```python
def countRec(a):
    f = open('Book.dat', 'rb')
    while True:
        try:
            data = pickle.load(f)
            if data[2] == a:
                print(data)
        except:
            break
countRec('mno')
```

```python
''' Given a binary file game.dat, containing records of following list format: [game_name, participants]
 Write a function in Python that would read contents from the file game.dat and 
 creates a file named basket.dat copying only those records from game.dat where the game name is "Basket Ball"'''
```

```python
# creating file first
import pickle
while True:
    with open('game.dat', 'ab') as f:
        game_name = input('enter game name: ')
        particp = input('enter participate names: ')
        rec= [game_name, particp]
        pickle.dump(rec, f)
        choice = input('want to add more in record y/n: ')
        if choice == 'n':
            break
            
                                            # now solving
def copy(game_name):
    out_file = open('basket.dat', 'ab')
    with open('game.dat', 'r+b') as f:
        while True:
            try:
                rec = pickle.load(f)
                if game_name == rec[0]:                           #  NOT GETING ANSWER --PLZ SUGGEST HERE ONLY
                    pickle.dump(rec, out_file)
            except:
                break
    out_file.close()
game_name= input('enter game name to be searched: ')
copy(game_name)
```
